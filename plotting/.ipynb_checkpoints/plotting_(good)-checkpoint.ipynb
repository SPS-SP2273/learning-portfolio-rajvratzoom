# 2 Comparing the two ‘dialects’

**pyplot vs OO dilects:**

````python
from matplotlib import pyplot as plt
import numpy as np
#data
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
````

### pyplot Interface

````python
#what we learnt previously
plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()
````

### OO Interface

````python
# an example
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, cos_x, label='cos x')
ax.plot(x, sin_x, label='sin x')
ax.legend()
````

We first start by asking Matplotlib to create a figure and axis using ```subplots()```.

This will give us a figure ```fig``` and an axis ```ax```.(You can assign any name to the figure and axis varibles)

We will be plotting a single column and row below

OO seems more complicated to code but it gives one more versatility. A general tip is to use pyplot for quick and unrefined plots while the OO interface can be used for mroe complex plots that need require more freedom and need to look more refined.

# 3 What is OO ax, really?

The code below creates the plot shown. Let's break it down

````python
fig, ax = plt.subplots(nrows=2, ncols=1)

ax[0].plot(x, cos_x, label='cos x')
ax[1].plot(x, sin_x, label='sin x')

ax[0].legend()
ax[1].legend()
````

We must first ask for 2 rows and 1 column, just as how the figures are arranged in the plot above. The code for this is shown below:

````python
fig, ax = plt.subplots(ncols=1, nrows=2)
````

````python
type(ax) #ax is a Numpy array
````

````python
ax.shape #ax has 2 'things' inside
````

````python
type(ax[0]) #it contains a matplotlib axis!
````

# 4 A complete OO example

A walkthough of OO code

````python
fig, ax = plt.subplots(nrows=2, ncols=1,           #get 2 axes ready to plot in two rows in one column
                       figsize=(5, 5),             #change the figure size
                       sharex=True)                #ask that the plots share the x axis

#note that we can index the axes in ax using 0 and 1.
ax[0].plot(x, cos_x, label=r'$\cos(x)$')           #then we just .plot to that axis, the axis label goes here as well
ax[0].fill_between(x, 0, cos_x, alpha=.25)         #use fill_between() to fill shade between the plots
ax[1].plot(x, sin_x, label=r'$\sin(x)$')           #same 2 instructions as above for the second axis
#ax[1].fill_between(x, 0, sin_x, alpha=.25)        #look at graph for example without shade fill

for a in ax:                                       #for loop to iterate through the items in ax instaed of using indexing                 
    a.legend()                                     #responsible for plotting the legends
    a.grid(alpha=.25)                              #setting a grid to each plot
    a.set_ylabel('$y$', rotation=0)                #setting the y-axis labels

ax[1].set_xlabel('$x$')                            #sharing the x axis so only need to label the lowest plot

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')           #adding a title
fig.tight_layout()                                 #final edits by matplotlib to make the figure look nice
````

````python
#WITHOUT fig.tight_layout()   # graphs are just abit further from each other now
fig, ax = plt.subplots(nrows=2, ncols=1,           #get 2 axes ready to plot in two rows in one column
                       figsize=(5, 5),             #change the figure size
                       sharex=True)                #ask that the plots share the x axis

#note that we can index the axes in ax using 0 and 1.
ax[0].plot(x, cos_x, label=r'$\cos(x)$')           #then we just .plot to that axis
ax[0].fill_between(x, 0, cos_x, alpha=.25)         #use fill_between() to fill the plots
ax[1].plot(x, sin_x, label=r'$\sin(x)$')           #same 2 instructions as above for the second axis
ax[1].fill_between(x, 0, sin_x, alpha=.25)

for a in ax:                                       #for loop to iterate through the items in ax instaed of using indexing                 
    a.legend()                                     #like the previous example
    a.grid(alpha=.25)                              #setting a grid to each plot
    a.set_ylabel('$y$', rotation=0)                #setting the labels

ax[1].set_xlabel('$x$')                            #sharing the x axis so only need to label the lowest plot

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')           #adding a title
````

# 5 Multiple rows and columns

We will now cover how to work with multiple rows and columns that form a grid of plots like the colorful one below.

![pic](https://sps.nus.edu.sg/sp2273/docs/python_basics/07_plotting/__imgs__/plotting_good_exercise-01_solution.png)

````python
We will first start by making the graph below:
````

![pic](https://sps.nus.edu.sg/sp2273/docs/python_basics/07_plotting/__imgs__/plotting_good_multiple_main.png)

````python
#final code we wish to create
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5),
                       sharex='col', sharey='col')

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
````

## 5.1 Sharing axes

````python
#WITHOUT sharex and sharey
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5)
)

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
````

It can be seent that the plots have been auto-scaled without aligning the scale with regard to any other figure as previously seen

Here are some ways to specify how to share the axes:

| Option | Result                                  |
|--------|-----------------------------------------|
| `True` | Makes all the axes use the same range.  |
| `col`  | Use the same range for all the columns  |
| `row`  | Use the same range for all the rows     |


````python
#sharing x axis scale
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5),
    sharex=True, sharey='row'
)

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
````

````python
# sharing y axis scale
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5),
    sharex='col', sharey='row'
)

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
````

## Accessing all Axes

To make changes to all axes, we could refer to them using the code below:

````python
top_left.grid(alpha=.25)
top_right.grid(alpha=.25)
bottom_left.grid(alpha=.25)
bottom_right.grid(alpha=.25)
````

A more efficient way to loop through the axes instead:

````python
for a in ax.flatten():
    a.grid(alpha=.25)
````

## 6.1 Histograms

A histogram can be useful to visualise a distribution within data. We will be using data from [sg.gov](https://data.gov.sg/) regarding the mean monthly earnings of graduates from various universities in Singapore

### Data

I have already downloaded the data files into the repository!

### A quick helper function

We will be reading data from the data files multiple times. In order to speed things up, we will be making use of a helper function: ```det_plot_data()```. Also take note that we will be skipping the first row of the file as the first row are simply the labels of the columns as seen when we open the files to examine them.

````python
#helper function
def get_plot_data():
    data = {}
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
    data['All'] = np.loadtxt(filename, skiprows=1)

    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
    data['NUS'] = np.loadtxt(filename, skiprows=1)

    return data
````

### The histogram

````python
plt.style.use('bmh')  #'bmh' style has been used here
data = get_plot_data()

# bins specifies how many bins to split the data
plt.hist([data['All'], data['NUS']], bins=50, label=['All', 'NUS'])
plt.xlabel('Mean of Basic Montly Earning (S$)')
plt.ylabel('Number of Students')
plt.legend()
````

## 6.2 Scatter plots

Scatter plots are created by putting a marker at an (x,y) point you specify. Over here we will need to specify x values, so we will use ```range()``` along with ```len()``` to generate a list to fit the data

````python
plt.style.use("seaborn-v0_8-darkgrid") #'seaborn-v0_8-darkgrid' syle has been used here

data = get_plot_data()

for label, numbers in data.items():
    x = range(len(numbers))        #The function used to generate x values
    y = numbers
    plt.scatter(x, y, label=label, alpha=.5)

plt.xlabel('Position in the list')
plt.ylabel('Mean of Basic Montly Eraning (S$)')
plt.legend()
````

## 6.3 Bar charts

 Dummy data has been entered here. The data was converted to lists and made into a bar graph. Moreover, the barh() can be used for horizontal bars:

````python
student_numbers = {'Life Sciences': 14,   #dummy data
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

plt.style.use('ggplot')  # 'ggplot' style is used
plt.bar(majors, numbers)
plt.xlabel('Majors')
plt.ylabel('Number of Students')
````

````python
#HORZIONTAL BAR GRAPH
student_numbers = {'Life Sciences': 14,   #dummy data
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

plt.style.use('ggplot')  # 'ggplot' style is used
plt.barh(majors, numbers)
plt.xlabel('Majors')
plt.ylabel('Number of Students')
````

## 6.4 Pie charts

````python
student_numbers = {'Life Sciences': 14,  #dummy data
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

plt.style.use('fivethirtyeight')  #fivethirtyeight style
plt.pie(numbers, 
        labels=majors,
        autopct='%1.1f%%',   # How to format the percentages
        startangle=-90                
        )
plt.title('Percentage of each major')
````
