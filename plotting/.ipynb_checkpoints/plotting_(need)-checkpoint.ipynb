# 1 A simple plot

## 1.1 Letâ€™s look at some code.

````python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='blue', linestyle='dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.90)
plt.legend(loc='lower right')
````

### Things to note

1. Alternative abbreviations:

| Long form  | Abbreviation |
|------------|--------------|
| color      | c            |
| linestyle  | ls           |
| linewidth  | lw           |


both plots below will create the same line:

````python
plt.plot(x, y_1, color='red', linestyle='dashed', linewidth=2)
````

````python
plt.plot(x, y_1, c='red', ls='dashed', lw=2)
````

2. Jupyter is an interactive environment, so you will see an output even if you omit ```plt.show()```. However, it is good practice to include this line anyway so your code will also work in non-interactive environments (for instance, when the script is run directly from the command line).

3. There are default styling parameters so one does not necessarily have to define all the parameters. An example is shown below:

````python
plt.plot(x, y_1, y_2)
````

4. You can split the arguments into separate lines to improve readability.
So, both of the following forms below are acceptable:

````python
plt.plot(x, y_1, color='red', linestyle='dashed', linewidth=2)
````

````python
plt.plot(x, y_1, label='Y values',
           color='red', linestyle='dashed')
````

5. The order of how you specify the keyword arguments (color, linewidth, â€¦) does not matter.

# 2 Better with NumPy

````python
import numpy as np
````

We will now redo the plots without the styling but take note that styling is important to maintain professionalism and quality of the plot. Moreover, we will also be using NumPy arrays instead of python lists from now on as there are far more benefits to using it.

````python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
````

````python
## 2.1 Adding mathematical functions
````

````python
One of the advantages of NumPy arrays is that they allow us to generate data-related mathematical functions easily. Letâ€™s reuse our previous code to plot 
 and 
````

````python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
````

The plot does not look nice as the x^2 value continues to increase while the sin(x) max value is 1. To address this we will add another axis

## We need another axis!

The simplest way is to have another y-axis that shares the same x-axis. We can use the command ```twinx()``` for this.

````python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')     

#take note that there are 2 legend calls now!
````

### Things to note

The plot is not smooth as we have very few points. We will use ```np.linspace()``` to generate more x values to make the curve smoother

````python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')     
````

# 3 Saving to disc

To use the plot elsewhere like in a presentation, we must first save it to disk. Matplotlib allows us to save the plot in many formats. Here are some instructions below:

- we will be using ```savefig()```
- we then specify the format with the extension of the filename(e.g. ```filename.pdf```)
- Then we specify the resolution using dpi

Doing this will simply save the plot in the working directory but you can specify to save it elsewhere as well using the code below:

````python
plt.savefig('C://Desktop/simple-plot.png', dpi=150)  #location first then decide the name and extension
````

I can save the file in other formats lik JPEG using ```simple-plot.jpeg``` or increase the dpi if I wanted a higher resolution

# 4 A real example: Global Warming

## 4.1 Plotting data from files

````python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')  #adds a horziontal line at the 0 value to allow us to visualise the trend better
````

## 4.2 Styles

you can also use preset styles from this website: [style options](http://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html) This is so that we will not have to keep configuring the styling parameters everytime we want to plot a new graph

````python
#this line must be added right at the top to configure all graphs
plt.style.use('dark_background')  # I like black haha
````

````python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
````

````python
#code for the final product

plt.style.use('classic')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]


plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='red')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
plt.show()
````
