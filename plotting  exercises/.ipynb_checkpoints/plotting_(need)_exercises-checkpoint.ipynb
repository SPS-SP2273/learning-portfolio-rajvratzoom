# Exercise 1

```python
import matplotlib.pyplot as plt
```

```python



# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y,
             xerr=x_err, yerr=y_err,
             color='r', linestyle='--',
             marker='o', markerfacecolor='yellow',
             ecolor='k', capsize=5,
             label='Y values with errors')

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

# Exercise 2

```python
import numpy as np
import matplotlib.pyplot as plt


x_values = np.linspace(0, 50, 500)  # 500 points from 0 to 50
y_values = np.exp(-x_values / 10) * np.sin(x_values)

plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, label=r'$y = e^{-\frac{x}{10}} \sin(x)$')


plt.xlabel('x', fontsize=15, color='darkorange')  #dark orange x and y labels with fontsize 15
plt.ylabel('y', fontsize=15, color='darkorange')
plt.title(r'$y = e^{-\frac{x}{10}} \sin(x)$', fontsize=20) #equation in title
plt.grid(True, alpha=0.25) #0.25 opacity
plt.ylim([-1, 1]) #y limits

plt.show()
```
